```python
import torch
import numpy as np
```

```python
#1D tensor
t1 = torch.tensor([1, 2, 3])
print("1D:\n",t1)

#2D tensor
t2 = torch.tensor([[1, 2, 3],[4, 5, 6]])
print("\n2D:\n",t2)

#3D tensor
t3 = torch.randn(2, 2, 3)
print("\n3D:\n",t3)
```

```python
#basic operations
a = torch.tensor([1, 2, 3])
b = torch.tensor([4, 5, 6])
print("Add:",a + b)
print("Subtract:",a-b)
print("Multiply:",a*b)
print("Divide:",a/b)
```

```python
#indexing and slicing
x = torch.tensor([[10, 20, 30],[40, 50, 60],[70, 80, 90]])
print("First row:",x[0])
print("Column 1:",x[:,1])
print("Slice rows 0-1 and cols 1-2:\n", x[0:2,1:3])
```

```python
#boolean masking
y = torch.tensor([1, 4, 7, 2, 9, 3])
mask = y > 3
print("Mask:",mask)
print("Values>3:", y[mask])
```

```python
#extracting subtensors
matrix = torch.arange(1, 17).reshape(4, 4)
print("Original:\n", matrix)
print("\nTop left 2x2:\n",matrix[:2, :2])
print("\nBottom row:\n",matrix[-1])
```

```python
#functions
t = torch.arange(1, 7)
print("Original:",t)

#reshape
print("Reshape to 2x3:\n", t.reshape(2, 3))

#view
print("View to 3x2:\n", t.view(3, 2))

# unsqueeze
print("Unsqueeze(add dim at 0):", t.unsqueeze(0))
print("Unsqueeze(add dim at 1):", t.unsqueeze(1))

#squeeze
ts = torch.tensor([[1, 2, 3]])
print("Squeeze:", ts.squeeze())
```

```python
#comparing .reshape()
arr = np.arange(1, 7)
print("NumPy reshape:\n", arr.reshape(2, 3))
ten = torch.arange(1, 7)
print("PyTorch reshape:\n", ten.reshape(2, 3))
```

```python
#broadcasting
A = torch.tensor([[1, 2, 3],
                  [4, 5, 6]])
b = torch.tensor([10, 20, 30])
print("Broadcasted result:\n",A+b)
```

```python
#in place vs out of place operations
a = torch.tensor([1, 2, 3])
b = a + 5
print("Original:", a)
print("New:", b)
```
